```python
#Required Libraries for importing the data
import pandas as pd

#Required Libraries for decision tree visualization
from matplotlib import pyplot as plt
from sklearn import tree

#Required Libraries for training and testing the data using Decision Tree Classifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

#Required Libraries for Model Evaluation
from sklearn import metrics
```

# Importing the data

```python
df = pd.read_csv("Iris.csv")
```

# Training and Testing the Algorithm

```python
#Putting independent and dependent variables into X and Y variable respectively
X = df.drop('Species', axis = 1)
Y = df.Species
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.33, random_state=1)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

Computing the accuracy of the algorithm

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

# Decision Tree Visualization

```python
b = plt.subplots(nrows = 1, ncols = 1, figsize = (5,5), dpi = 200 )
tree.plot_tree(clf);
```

```python
feature_names=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
species_names=['setosa', 'versicolor', 'virginica']

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=400)

tree.plot_tree(clf,
           feature_names = feature_names, 
           class_names=species_names,
           filled = True);

fig.savefig('imagename.png')
```
